## Problem 1:  Controls

Write a Python script that proves that the lines of data in Germplasm.tsv, and LocusGene are in the same sequence, based on the AGI Locus Code (ATxGxxxxxx).  (hint: This will help you decide how to load the data into the database)

```python
#First we need to import the files we are going to use and then open them and read their lines

import re

locus_file = open ("LocusGene.tsv", "r")
germ_file = open ("Germplasm.tsv", "r")

locus_file = locus_file.readlines()
germ_file = germ_file.readlines() 

#we can now check if the size of both files is the same

locus_size = len(locus_file)
germ_size = len(germ_file)

if locus_size == germ_size:
    print("both files have the same size")
else:
    print("the files don't have the same size")
    

#Now, we know that both files have the same size but we need to know if the AGI Locus code in both of them match. For this, we can create a loop in which it takes all the AGI LOcus Code from both files and copy them into lists in which we can check if the match or not.

locus = []
germ = []

for i in range (0,locus_size):
    matchObj = re.search(r'AT\dG\d+', locus_file[i])
    matchObj_2 = re.search(r'AT\dG\d+', germ_file[i])
    if matchObj:
        Locus_AGI = matchObj.group()
        locus.append(Locus_AGI) 
        
        Germ_AGI = matchObj_2.group()
        germ.append(Germ_AGI)
        
    else:
        print("Done")

if locus == germ:
    print("Both files are the same")
else:
    print ("The files are not the same")
    
print (locus)
print ("")
print(germ) 
```

## Problem 3: Fill the database
Using pymysql, create a Python script that reads the data from these files, and fills the database.  There are a variety of strategies to accomplish this.  I will give all strategies equal credit - do whichever one you are most confident with.

```python
import pymysql.cursors

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='Exam_2',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

connection.autocommit(True)

try:
    with connection.cursor() as cursor:
        for i in range (1,len(locus_file)):
            locus_filled = locus_file[i].split('\t')
            
            locus = locus_filled [0]
            gene = locus_filled[1]
            Protein_length = locus_filled[2]
            
            sql = "INSERT INTO locus_table(locus, gene, Protein_length) values ('{}', '{}', '{}')".format(locus, gene, Protein_length)
            cursor.execute(sql)
            
            
            germ_filled = germ_file[i].split('\t')
            #in this table we begin with [1] because both tables have the column "locus" and it is the same in both
            germplasm = germ_filled [1]
            phenotype = germ_filled [2]
            pubmed_link = germ_filled [3]
            
            sql = "INSERT INTO germ_table(locus, germplasm, phenotype, pubmed_link) values ('{}', '{}', '{}', '{}')".format(locus, germplasm, phenotype, pubmed_link)
            cursor.execute(sql)
            
            connection.commit()
            
finally:
    print("Everything correct")
    connection.close()

    
#In this exercise, in order to read the files and fill the tables created, we can use pymysql to extract the data from the files using "for"and then tell the program to fill the tables with the data from the files using "try-with" giving the names of the columns and the format we need to fill

```

## Problem 4: Create reports, written to a file

1. Create a report that shows the full, joined, content of the two database tables (including a header line)

2. Create a joined report that only includes the Genes SKOR and MAA3

3. Create a report that counts the number of entries for each Chromosome (AT1Gxxxxxx to AT5Gxxxxxxx)

4. Create a report that shows the average protein length for the genes on each Chromosome (AT1Gxxxxxx to AT5Gxxxxxxx)

When creating reports 2 and 3, remember the "Don't Repeat Yourself" rule! 

All reports should be written to **the same file**.  You may name the file anything you wish.

```python
import pymysql.cursors
import csv

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='Exam_2',
                             charset='utf8mb4',  
                             cursorclass=pymysql.cursors.DictCursor)

connection.autocommit(True)

try:
    with connection.cursor() as cursor:
        sql = "SELECT * FROM germ_table, locus_table WHERE germ_table.locus = locus_table.locus"
        cursor.execute(sql)
        results = cursor.fetchall()
        with open ('AIDA_FINAL_REPORT.csv', 'w') as csvfile:
            answerwriter = csv.writer(csvfile, delimiter="\t", quotechar='"')
            answerwriter.writerow(["Problem 1"])
            answerwriter.writerow(["Locus", "Germplasm", "Phenotype", "pubmed_link", "Gene", "Protein_length"])
            for line in results:
                answerwriter.writerow([line['locus'], line['germplasm'], line['phenotype'], line['pubmed_link'], line['gene'], line['Protein_length']])

    connection.commit() 
        
finally:
    print("Problem 1 done")
    connection.close()
    

l = open('AIDA_FINAL_REPORT.csv', 'r') ## In order to see the content of the file every time we add some information. 
for line in l.readlines():
    print(line)
    
```

```python
#In this problem we need to count the number of entries on each chromosome. we will follow the same estructure but using SELECT COUNT and using regular expresion in order to query for the chromosomes. 
# we can set i = 1 at the beginning and tell the program to catch all the answers while i <= 5 until i =1. In this way we are cerating a loop that reads the table looking for each chromosome and recording each time it finds it. 
import csv
import pymysql.cursors

i = 1

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='Exam_2',
                             charset='utf8mb4',  
                             cursorclass=pymysql.cursors.DictCursor)

connection.autocommit(True)

try:
    with connection.cursor() as cursor:
        
        with open('AIDA_FINAL_REPORT.csv', 'w', newline='') as csvfile:
            answerwriter = csv.writer(csvfile, delimiter="\t", quotechar='"')
            answerwriter.writerow(["Problem 3"])
            answerwriter.writerow(["Number of entries for each chromosome"])
            while i <= 5:
                sql = "SELECT COUNT(locus) FROM germ_table WHERE locus REGEXP (%s)"
                data = ("^AT"+str(i)+"G.+") 
                cursor.execute(sql, data)
                results = cursor.fetchall()
                for line in results:
                    answerwriter.writerow(["AT"+str(i)+"G", line['COUNT(locus)']])
                i += 1
                
        connection.commit() 
        
finally:
    print("problem 3 done")
    connection.close()
```

```python
# In the same way, we can repeat this structure of code in order to show the average protein length for the genes on each Chromosome
import csv
import pymysql.cursors

i = 1

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='Exam_2',
                             charset='utf8mb4',  
                             cursorclass=pymysql.cursors.DictCursor)

connection.autocommit(True)

try:
    with connection.cursor() as cursor:
        
        with open('AIDA_FINAL_REPORT.csv', 'w', newline='') as csvfile:
            answerwriter = csv.writer(csvfile, delimiter="\t", quotechar='"')
            answerwriter.writerow(["Problem 4"])
            answerwriter.writerow(["Average protein lenght for each chromosome"])
            while i <= 5:
                sql = "SELECT ROUND(AVG(Protein_length)) FROM locus_table WHERE locus REGEXP (%s)"
                data = ("^AT"+str(i)+"G.+") 
                cursor.execute(sql, data)
                results = cursor.fetchall()
                for line in results:
                    answerwriter.writerow(["AT"+str(i)+"G", line['ROUND(AVG(Protein_length))']])
                i += 1
                
        connection.commit() 
        
finally:
    print("problem 4 done")
    connection.close()
    
    

l = open('AIDA_FINAL_REPORT.csv', 'r')
for line in l.readlines():
    print(line)
```
